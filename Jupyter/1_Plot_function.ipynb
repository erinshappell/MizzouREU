# Get started with Math
### Plotting a function
Plot a straight line

```python
import numpy as np # A generic library for arrays and matrices
import matplotlib.pyplot as plt # library for plotting
# to enable plotting in line in notebook
%matplotlib inline

x = np.arange(-5,5,0.5) # create an array of x from -10 to 10 with step 1
print(x)

# define a function for straight line, given x, slope m, y-intersept b
def line(x,m,b):
    y = m*x+b
    return y
```

```python
m = 1
b = 0
y = line(x,m,b)
print(y)

plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.grid()
plt.show()
```

Plot a parabola

```python
# define a function for parabola function, given x, in vertex form
def parabola(x,xv,yv,a):
    y = a*(x-xv)**2+yv
    return y

xv = 0
yv = 0
a = 1
y = parabola(x,xv,yv,a)

plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.grid()
plt.show()
```

Approximate the tangent line and the slope at some point

```python
idx1 = 10 # index of a point in x
d = 5 # increment of index

idx2 = idx1+d # index of a second point in x
x1,x2 = x[idx1],x[idx2]
print([x1,x2])

P,Q = [x1,parabola(x1,xv,yv,a)],[x2,parabola(x2,xv,yv,a)] # two points on the parabola
print([P,Q])

# define a function for a straight line determined by two points
def line2pts(x,P0,P1):
    y = (x-P0[0])*(P1[1]-P0[1])/(P1[0]-P0[0])+P0[1]
    return y

plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.plot((P[0],Q[0]),(P[1],Q[1]),'r.')
plt.plot(x,line2pts(x,P,Q),'g')
plt.grid()
plt.show()
```

Plot the curves of exponetial function and sine function using function below. And approximate a tangent line for each.

```python
# Get tangent line at two points
idx1 = 10 # index of a point in x
d = 4 # increment of index

idx2 = idx1 + d # index of a second point in x
x1,x2 = x[idx1],x[idx2]

P,Q = [x1,np.exp(x1)],[x2,np.exp(x2)] # two points on the graph

# Plot exponential function with tangent line and points
y = np.exp(x)
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.plot((P[0],Q[0]),(P[1],Q[1]),'r.')
plt.plot(x,line2pts(x,P,Q),'g')
plt.grid()
plt.show()
```

```python
# Get tangent line at two points
idx1 = 10 # index of a point in x
d = 2 # increment of index

idx2 = idx1 + d # index of a second point in x
x1,x2 = x[idx1],x[idx2]

P,Q = [x1,np.sin(x1)],[x2,np.sin(x2)] # two points on the graph

# Plot exponential function with tangent line and points
y = np.sin(x)
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.plot((P[0],Q[0]),(P[1],Q[1]),'r.')
plt.plot(x,line2pts(x,P,Q),'g')
plt.grid()
plt.show()
```
