# Introduction to Plotting in Python
This notebook will walk you through the basics of using Python's libraries to create and plot basic functions and tangent lines!
In high school, you probably learned about how to plot and analyze basic functions like lines, parabolas, exponentials, and trigonometic functions. While a TI-89 works fine for plotting these, it can be useful to know how to use Python to customize and plot these functions as well!
First, let's walk through how to create and plot a linear function.

```python
# First let's import some libraries...
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import YouTubeVideo
```

Before we begin, it may be helpful to refresh yourself linear equations and how to plot them. Run the code below to reveal a Khan Academy video on slope-intercept form.

```python
# Now, let's define a function for creating the points of our line
def create_line(x,m,b):
    y = m * x + b
    return y

# Now, let's create the line y = 2x + 1
x = np.arange(-10,11) # Using the range -10 < x < 10
slope = 2
y_int = 1
y = create_line(x,slope,y_int) # y = 2x + 1

# Now, let's plot the line
plt.figure()
plt.grid()
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plotting a Line')
plt.show()
```

Now, let's spice things up a bit--let's add some __points__ to our line.

```python
# Create a point in the range -10 < x < 10
x_pt = 0;
y_pt = create_line(x_pt,slope,y_int)

# Plot the point, along with the original function, on a graph
plt.figure()
plt.grid()
plt.plot(x,y)
plt.plot(x_pt,y_pt,'r.') # Plot the point as a red dot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plotting a Line')
plt.show()
```

Great! Now, let's look at plotting a __parabola__. First, run the code below to view a Khan Academy video on parabolic functions to refresh your knowledge of them.

```python
YouTubeVideo('7QMoNY6FzvM')
```

```python
# Now, let's define a function for creating the points on our parabola
def create_para(x,a,n,xv,yv):
    y = a * (x - xv) ** n + yv
    return y

# Now, lets create the parabola 3(x - 2)^2 + 1
a = 3
n = 2
xv = 2
yv = 1
y = create_para(x,a,n,xv,yv)

# Now, let's plot the line
plt.figure()
plt.grid()
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plotting a Parabola')
plt.show()
```

You may have noticed that, in the code above, we did not __define__ x like we did when we created the line. This is because we already __defined__ x in another code block, so the kernel already has a __value__ assigned to x (the numbers in the range (-10,10)).

Once again, let's plot some points on the graph.

```python
# Create two points in the range -10 < x < 10
x_pt1 = -2
dx = 3
x_pt2 = x_pt1 + dx
y_pt1 = create_para(x_pt1,a,n,xv,yv)
y_pt2 = create_para(x_pt2,a,n,xv,yv)

# Plot the points, along with the original function, on a graph
plt.figure()
plt.grid()
plt.plot(x,y)
plt.plot(x_pt1,y_pt1,'r.') # Plot the points as red dots
plt.plot(x_pt2,y_pt2,'r.') 
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plotting a Parabola with Two Points')
plt.show()
```

Now that we have two points plotted on our parabola, we can estimate the derivative at the first point by plotting the __secant line__ going through the two points. This estimation will become more accurate (i.e. the secant line will converge to the __tangent line__ through the first point). This procedure is shown in the figure below:

![SlopeUrl](https://i.stack.imgur.com/3snPJ.gif "slope")

Now, let's plot the __secant line__ through those two points on our parabola.

```python
# Function for finding the equation of a line through two points
def line2pts(x,P0,P1):
    y = (x-P0[0])*(P1[1]-P0[1])/(P1[0]-P0[0])+P0[1]
    return y

# Plot the functions and points
plt.figure()
plt.grid()
plt.plot(x,y)
plt.plot(x_pt1,y_pt1,'r.') # Plot the points as red dots
plt.plot(x_pt2,y_pt2,'r.') 
plt.plot(x,line2pts(x,(x_pt1,y_pt1),(x_pt2,y_pt2)),'g') # Plot the secant line as a green line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plotting a Parabola and a Secant Line through Two Points')
plt.show()
```

Take a moment to return to the code block where __dx__ was defined; try increasing and decreasing the value of dx. What happens to the secant line? 

Now, let's plot the __derivative__ of our parabola. The __derivative__ of a parabola is a line, and each value represents the change in y over time at a specific point in time.

```python
# Function for finding the derivative of a parabola
def derive_para(x,a,n,xv):
    dy_dx = n * a * (x - xv)
    return dy_dx

dy_dx = derive_para(x,a,n,xv)

# Plot the functions and points
plt.figure()
plt.grid()
plt.plot(x,y)
plt.plot(x,dy_dx,'o') # Plot the derivative of the parabola as an orange line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plotting a Parabola and its Derivative')
plt.show()
```

Cool! What about the __integral__ of a parabola? Taking the integral of a parabola may be viewed as the opposite of taking the derivative; so the integral of our parabola is going to be a __cubic__ function.

```python
# Function for finding the integral of a parabola
def integrate_para(x,a,n,xv,yv):
    Y = (a / (n + 1)) * (x - xv) ** (n + 1) + yv * x
    return Y

Y = integrate_para(x,a,n,xv,yv)

# Plot the functions and points
plt.figure()
plt.grid()
plt.plot(x,y)
plt.plot(x,Y,'g') # Plot the integral of the parabola as an green line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plotting a Parabola and its Derivative')
plt.show()
```

The integral of our parabola represents the area under the parabola at each point in time. The formula for integration is derived from a __Riemonn Sum__ that begins by estimating the area under a curve by dividing the area into rectangles of width __w__, then shrinking w so that the rectangles are infintesimally small.
